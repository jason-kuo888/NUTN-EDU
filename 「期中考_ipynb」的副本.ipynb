```
install.packages("SportsAnalytics")    #安裝SportsAnalytics
library("SportsAnalytics")           #載入套件  
NBA1819<-fetch_NBAPlayerStatistics("18-19")
```

```

NBA18190rder<-NBA1819[order(NBA1819$ThreesMade,decreasing=T),]     #逗號前方放1~20,表示取1~20列;逗號後方空白,表示要取所有欄位
NBA18190rder[1:20]
```

```
subset(NBA1819,TEAM=LAL)  #subset()欄位條件搜尋
```

```
NBA1819[grepl("Lebron James",NBA1819$Name),]   #用grepl()作字串條件蒐尋Lebron James
```

```

install.packages("ggplot2") #安裝ggplot2
library(ggplot2)      #讀取ggplot2
```

```
NBA1819<-fetch_NBAPlayerStatistics("18-19") #讀入資料
qplot(FieldGoalsAttempted, TotalPoints, data=NBA1819) #投籃出手次數與總得分
```

```
qplot(ThreesMade, TotalPoints, data = NBA1819,
geom = c("point", "smooth"))               #可在圖上加點與漸進線,如使用geom = c("point", "smooth")   (X=三分球,Y=分數)
```

```

qplot(FieldGoalsAttempted, TotalPoints, data=NBA1819, color=Position) #守備位置著色  (X=FieldGoalsAttempted,Y=TotalPoints)
```

```
str(NBA1819) #確認資料型別
qplot(FieldGoalsAttempted, TotalPoints, data=NBA1819, color=Team) #隊伍著色      (X=FieldGoalsAttempted,Y=TotalPoints)
```

```
qplot(TotalPoints, data=NBA1819, fill=Team) #輸入的變量為單變量而非雙變量時，預設圖形會從散佈圖變直方圖
```

```
qplot(TotalPoints, data=NBA1819,
   facets=Position ~ . , biwidth=10) #每10分一組
```

```
df<-data.frame(gp=factor(rep(letters[1:3],each=10)),y=rnorm(30))
ggplot(df,aes(x=gp,y=y))+geom_point()                   #產生範例用的畫圖資料
```

```
ggplot(df,aes(x=gp,y=y))+geom_boxplot()                 #用geom_boxplot()改畫盒狀圖 
```
