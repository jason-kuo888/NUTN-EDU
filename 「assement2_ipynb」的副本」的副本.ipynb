```
install.packages("SportsAnalytics") 
install.packages("ggplot2") 

library(SportsAnalytics)
library(ggplot2) 
```

```
NBA1819<-fetch_NBAPlayerStatistics("18-19") #讀入資料
qplot(FieldGoalsAttempted, TotalPoints, data=NBA1819) #投籃出手次數與總得分
```

```
qplot(FieldGoalsAttempted, TotalPoints, data=NBA1819, color=Position) #守備位置著色
```

```
str(NBA1819) #確認資料型別
qplot(FieldGoalsAttempted, TotalPoints, data=NBA1819, color=Team) #隊伍著色
```

```
qplot(TotalPoints, data=NBA1819, fill=Team) #輸入的變量為單變量而非雙變量時，預設圖形會從散佈圖變直方圖
```

```
qplot(TotalPoints, data=NBA1819,
   facets=Position ~ . , biwidth=10) #每10分一組
```
